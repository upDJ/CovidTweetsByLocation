```python
#!/usr/bin/env python
# coding: utf-8

# In[2]:

import datetime
import numpy as np
import pandas as pd

path = '/Users/deion/TweetsByLocation/covid19_tweets.csv'
df = pd.read_csv(path)

#pre null
nullcount = df.isnull().count()
print(nullcount)

#post null - removed all null entries

df = df.dropna(subset=['user_location'])
nullcount = df.isnull().count()
print(nullcount)

sortTypes = ['quicksort','mergesort','heapsort']
#quicksort, #mergesort, #heapsort

storeResults=dict()

for i in range(0,3):
    copydf = df
    start1 = datetime.datetime.now()
    df.sort_values(by=['user_location'],ascending=True,kind=sortTypes[i])
    finish1 = datetime.datetime.now()
    result1 = finish1 - start1
    storeResults[sortTypes[i]] = result1
print('\n\n',storeResults)
```
